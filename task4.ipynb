**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical $3 \times 3$ Prewitt filters:

```python
def prewitt_h(img):
    gx = numpy.zeros(img.shape)
    n = 1
    for p in numpy.ndindex(img.shape):
        if n <= p[0] < img.shape[0] - n and n <= p[1] < img.shape[1] - n:
            neighbourhood_sum = 0
            for q in numpy.ndindex(3, 3):
                if q[1] == 0:
                    neighbourhood_sum += (1/6)*img[p[0]-n+q[0], p[1]-n+q[1]]
                elif q[1] == 2:
                    neighbourhood_sum += (-1/6)*img[p[0]-n+q[0], p[1]-n+q[1]]
                elif q[1] == 1:
                    neighbourhood_sum += 0*img[p[0]-n+q[0], p[1]-n+q[1]]
            gx[p] = neighbourhood_sum
    return gx
    
def prewitt_v(img):
    gy = numpy.zeros(img.shape)
    n = 1
    for p in numpy.ndindex(img.shape):
        if n <= p[0] < img.shape[0] - n and n <= p[1] < img.shape[1] - n:
            neighbourhood_sum = 0
            for q in numpy.ndindex(3, 3):
                if q[0] == 0:
                    neighbourhood_sum += (1/6)*img[p[0]-n+q[0], p[1]-n+q[1]]
                elif q[0] == 2:
                    neighbourhood_sum += (-1/6)*img[p[0]-n+q[0], p[1]-n+q[1]]
                elif q[0] == 1:
                    neighbourhood_sum += 0*img[p[0]-n+q[0], p[1]-n+q[1]]
            gy[p] = neighbourhood_sum
    return gy
```
